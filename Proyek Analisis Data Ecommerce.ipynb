### Explore ...

**Tabel order_item dengan tabel product**

```python
merge_product_df = pd.merge(
        left = order_items_df,
        right = product_df,
        how = "left",
        left_on = "product_id",
        right_on = "product_id")

merge_product_df.head()
```

```python
most_sold_product = merge_product_df.groupby(by="product_category_name").order_id.nunique().sort_values(ascending=False).reset_index().head(10)
most_sold_product = most_sold_product.rename(columns={"order_id": "order"})
most_sold_product
```

```python
least_sold_product = merge_product_df.groupby(by="product_category_name").order_id.nunique().sort_values(ascending=True).reset_index().head(10)
least_sold_product = least_sold_product.rename(columns={"order_id": "order"})
least_sold_product
```

**Tabel sebelumnya dengan order review**

```python
merge_order_review_df = pd.merge(
        left = merge_product_df,
        right = order_reviews_df,
        how = "left",
        left_on = "order_id",
        right_on = "order_id")

merge_order_review_df.head()
```

```python
product_satisfaction = merge_order_review_df.groupby('product_category_name')['review_score'].mean().sort_values(ascending=False).reset_index().head(10)
product_satisfaction
```

```python
satisfied_customers = merge_order_review_df[merge_order_review_df['review_score'] >= 2]['order_id'].nunique()
total_customers = merge_order_review_df['order_id'].nunique()
satisfied_percentage = (satisfied_customers / total_customers) * 100
unsatisfied_customers = total_customers - satisfied_customers

satisfaction_tabel = pd.DataFrame({'Review':['Puas','Tidak Puas'],
                                   'Number Customer' : [satisfied_customers, unsatisfied_customers]})
print(satisfaction_tabel)
```

**Tabel merge sebelumnya dengan seller_df**

```python
merge_seller_df = pd.merge(
        left = merge_order_review_df,
        right = sellers_df,
        how = "left",
        left_on = "seller_id",
        right_on = "seller_id")

merge_seller_df.head()
```

```python
products_sold = merge_seller_df.groupby('seller_id')['product_id'].count().reset_index()
products_sold
```

```python
top_sellers = products_sold.sort_values('product_id', ascending=False).head(10)
top_sellers
```

```python
top_sellers_with_city = top_sellers.merge(merge_seller_df[['seller_id', 'seller_city']], on='seller_id', how='left')
top_sellers_with_city
```

```python
top_sellers_with_city.duplicated().sum()
```

```python
top_sellers_with_city = top_sellers_with_city.drop_duplicates()
top_sellers_with_city = top_sellers_with_city.rename(columns={"product_id": "product terjual"})
top_sellers_with_city
```

```python
seller_location = merge_seller_df.groupby(['seller_city', 'seller_state'])['product_id'].count().reset_index()
seller_location
```

```python
merge_payment_df = pd.merge(
        left = order_items_df,
        right = order_payment_df,
        how = "left",
        left_on = "order_id",
        right_on = "order_id")

merge_payment_df
```

```python
# Menghitung jumlah penggunaan setiap metode pembayaran
payment_counts = merge_payment_df['payment_type'].value_counts()
payment_counts
```
